```python
df = pd.read_csv("categories.csv", index_col=0)
df
```

```python
plt.scatter(df['x'], df['y'])
plt.show()
```

# Varianta 1 - le afisez pe toate, si le colorez in functie de coloana category

```python
plt.scatter(df['x'], df['y'], c="pink")
```

```python
df["color"] = df["category"].map({"a":"blue", "b":"red"})
df
```

```python
plt.scatter(df['x'], df['y'], c=df["color"]);
```

# Varianta 2 - fac 2 scatter plot pe aceeasi figura

```python
plt.scatter(df['x'], df['y'], c="pink")
```

```python
df_only_a = df[df["category"] == 'a']
df_only_b = df[df["category"] == 'b']

df_only_a
```

```python
df_only_b
```

```python

plt.scatter(df_only_a['x'], df_only_a['y'])
plt.scatter(df_only_b['x'], df_only_b['y']); # ia default 2 culori
```

```python

plt.scatter(df_only_a['x'], df_only_a['y'], c="blue") # culoarea este specificata
plt.scatter(df_only_b['x'], df_only_b['y'], c="red"); # culoarea este specificata
```

# Varianta 3 - folosesc pandas pentru a apela matplotlib

```python
df.plot.scatter("x", "y", c="pink");
```

```python
plt.scatter(df['x'], df['y'], c="pink");
plt.xlabel("Coordonata x")
plt.ylabel("Coordonata y");
```

```python
plt.scatter(df['x'], df['y'], c="pink");
plt.xlabel("Coordonata x")
plt.ylabel("Coordonata y")
plt.title("Grafic Scatter");
```

```python
plt.scatter(df['x'], df['y'], c="pink");
plt.xlabel("Coordonata x")
plt.ylabel("Coordonata y")
plt.grid();
plt.title("Grafic Scatter");
```

```python
plt.scatter(df['x'], df['y'], c="pink");
plt.xlabel("Coordonata x")
plt.ylabel("Coordonata y")
plt.grid(alpha=0.2);
plt.title("Grafic Scatter");
```

```python
plt.scatter(df_only_a['x'], df_only_a['y'], c="blue", label="Categoria a") # culoarea este specificata
plt.scatter(df_only_b['x'], df_only_b['y'], c="red", label="Categoria b") 

plt.xlabel("Coordonata x")
plt.ylabel("Coordonata y")
plt.grid(alpha=0.2);
plt.title("Grafic Scatter");
plt.legend();

```

```python
formatie = [
    [None] * 3 +  [titulari[10]] + [None] * 3, 
    [None, None, titulari[8],   None,  titulari[9], None, None],
    [None, titulari[5],   None,  titulari[6], None, titulari[7],  None],
    [titulari[1], None,  titulari[2], None, titulari[3], None, titulari[4]],
    [None] * 3 +  [titulari[0]] + [None] * 3 
]
print(formatie)
```

```python
fig, axes = plt.subplots(5, 7)

for index_row, row in enumerate(formatie):
    for index_col, col in enumerate(row):
        file_name = formatie[index_row][index_col]
        # print("file_name", file_name)
        if file_name:
            read_image = plt.imread(folder_name + "/" + file_name)
            axes[index_row, index_col].imshow(read_image)
        axes[index_row, index_col].axis("off")
plt.show()
```

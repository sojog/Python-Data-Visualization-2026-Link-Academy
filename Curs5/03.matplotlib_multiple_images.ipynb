```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np
```

# Matplotlib func based vs OOP based

## 1. Func Based

```python
x = [1, 2, 3]
y = [2, 4, 6]

plt.plot(x, y)
plt.title("Plot simplu")
plt.xlabel("Axa X")
plt.ylabel("Axa Y")
plt.show()
```

## 2. OOP Based

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel("Axa X")
ax.set_ylabel("Axa Y")
ax.set_title("Plot simplu")
plt.show()
```

# In Matplotlib pot crea mai multe subploturi

```python
plt.subplots(nrows=2, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0, 0].plot(x, y)
axes[1, 1].plot(y, x, c="red")
```

# 1. Barplot

```python
df["brand"].unique(), df.brand.nunique()
```

```python
over_15k_df = df[df["price"] > 15000]
over_15k_df.count()
```

```python
plt.rcParams["figure.figsize"] # schimbare generala
```

```python
plt.figure(figsize=(6.4*4, 4.8)) # schimbare specifica a lungimii imaginii
sns.barplot(df, x="brand", y="price")
plt.xticks(rotation=45);
```

```python
sns.barplot(over_15k_df, x="brand", y="price")
plt.xticks(rotation=45);
```

# 2. Count plot

```python
df.keys()
```

```python
sns.countplot(df, x="warrantyInformation")
plt.xticks(rotation=45);
```

```python
plt.rcParams["figure.figsize"] = [6.4 * 5, 4.8]

sns.countplot(df, x="brand")
plt.xticks(rotation=45);
```

```python
#plt.rcParams["figure.figsize"] #= [6.4, 4.8]

plt.figure(figsize=(6.4 * 2, 4.8))
sns.countplot(df, x="category")
plt.xticks(rotation=45);
```

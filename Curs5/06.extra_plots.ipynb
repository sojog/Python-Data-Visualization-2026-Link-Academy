```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np
```

```python
COUNT = 20
```

```python
grupuri = np.repeat(("A", "B", "C"),COUNT)
grupuri 
```

```python
valori_A = np.random.normal(5, 1, COUNT)
valori_B = np.random.normal(7, 1.5, COUNT)
valori_C =  np.random.normal(6, 0.5, COUNT)
```

```python
df = pd.DataFrame({
    "grup":grupuri,
    "valoare": np.concatenate([valori_A, valori_B, valori_C])

})
df.head()
```

# 1. Stripplot

```python
sns.stripplot(df, x="grup", y="valoare")
```

# 2. Swarmplot

```python
sns.swarmplot(df, x="grup", y="valoare")
```

# 3. Violinplot

```python
sns.violinplot(df, x="grup", y="valoare")
```

# Violin Plot vs boxplot

```python
sns.boxplot(df, x="grup", y="valoare")
```

```python
fig, axes = plt.subplots(2, 2)
sns.stripplot(df, x="grup", y="valoare", ax=axes[0, 0])
sns.swarmplot(df, x="grup", y="valoare", ax=axes[0, 1])
sns.violinplot(df, x="grup", y="valoare", ax=axes[1, 0])
sns.boxplot(df, x="grup", y="valoare", ax=axes[1, 1])
```

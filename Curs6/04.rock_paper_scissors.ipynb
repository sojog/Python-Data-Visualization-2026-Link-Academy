```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import matplotlib.animation as animation
from IPython.display import HTML
```

```python
OPTIONS = ["paper", "rock", "scissors"]
```

```python
OUTCOME = ["win", "draw", "loss"]
```

```python
def game_outcome (player, computer):
    if player == computer: return OUTCOME[1]

    if player == OPTIONS[0] :
        if computer == OPTIONS[2]: return OUTCOME[2]
        else:  return OUTCOME[0]

    if player == OPTIONS[1] :
        if computer == OPTIONS[0]: return OUTCOME[2]
        else:  return OUTCOME[0]
        
    if player == OPTIONS[2] :
        if computer == OPTIONS[1]: return OUTCOME[2]
        else:  return OUTCOME[0]
```

```python
df = pd.DataFrame(
    { "player": [], "computer":[], "outcome": []}
)
df
```

```python
games = []

while True:
    player = input("Introduceți o valoare (Paper, Rock şi Scissors )").lower()

    if player == "exit":
        break 
    
    while player not in OPTIONS:
        player = input("Introduceți o valoare (Paper, Rock şi Scissors )").lower()

    computer = str(np.random.choice(OPTIONS))
    print("Player: ", player, "Computer:", computer,  game_outcome (player, computer))

    games.append([player, computer, game_outcome (player, computer)])
```

```python
df = pd.DataFrame(games, columns=["player", "computer", "outcome"]
    
)
df.to_csv("games.csv")
```

```python
figure, axis = plt.subplots()


BAR_OPTIONS = ["player", "draw", "computer"]

def update_function(index):
    axis.clear()

    current_df = df[df.index <= index]
    values = current_df["outcome"].value_counts()    
    CURRENT_OUTCOME = [values.get("win", 0), values.get("draw", 0), values.get("loss", 0)]

    bars = axis.bar(BAR_OPTIONS, CURRENT_OUTCOME, color="skyblue")


    axis.set_ylabel("Games Played")
    axis.set_xlabel("Outcome")
    
    return bars

ani = animation.FuncAnimation(fig=figure, func=update_function, frames=df.index, repeat=True, interval=2000)
ani.save("games.gif", writer="ffmpeg")



plt.close();
HTML(ani.to_jshtml())
```

# Pot sa adaug label-uri la grafic

```python
x = [10, 100]
y = [20, 200]

plt.xlabel("Coordonata X")
plt.ylabel("Coordonata Y")

plt.scatter(x, y);
```

# Pot adauga si un titlu la grafic

```python
x = [10, 100]
y = [20, 200]

plt.title("Grafic scatter")

plt.xlabel("Coordonata X")
plt.ylabel("Coordonata Y")

plt.scatter(x, y);
```

# Graficul de tip "plot" - reprezintă o linie simplă

# Pe un grafic pot sa cuplez mai multe tipuri de reprezentare a numerelor

```python
x = [10, 100, 133]
y = [20, 200, 99]

plt.title("Grafic scatter")

plt.xlabel("Coordonata X")
plt.ylabel("Coordonata Y")

plt.scatter(x, y);
plt.plot(x, y);
```

# Pot sa colorez diferit aceste grafice

```python
x = [10, 100, 133]
y = [20, 200, 99]

plt.title("Grafic scatter")

plt.xlabel("Coordonata X")
plt.ylabel("Coordonata Y")

plt.scatter(x, y, c="red");
plt.plot(x, y);
```

```python
x = [10, 100, 133]
y = [20, 200, 99]

plt.title("Grafic scatter")

plt.xlabel("Coordonata X")
plt.ylabel("Coordonata Y")

plt.scatter(x, y, c="red");
plt.plot(x, y, c="green");
```

```python
x = [10, 100, 133]
y = [20, 200, 99]

plt.title("Grafic scatter")

plt.xlabel("Coordonata X")
plt.ylabel("Coordonata Y")

plt.scatter(x, y, c="red");
plt.plot(x, y, c="#F015F4");
```

# Matplotlib functioneaza cu "ArrayLike" adică și alte tipuri de secvențe

```python
x = range(5)
y = range(10, 15)

print("x=", list(x))
print("y=", list(y))

plt.scatter(x, y);
```

```python
x = [3, 3, 3, 3, 3]
y = range(5)

plt.scatter(x, y);
```

# Matplotlib este compatibil si cu numpy

```python
x = np.ones(5) * 3
y = np.arange(5)

print("x", x, type(x))
print("y", y, type(y))

plt.scatter(x, y);
```

# Imaginea generata poate fi salvata programatic

```python
x = np.ones(5) * 3
y = np.arange(5)

print("x", x, type(x))
print("y", y, type(y))

plt.scatter(x, y)

plt.savefig("01.imagine_salvata_programatic.png");
```

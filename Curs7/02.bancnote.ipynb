```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
def calculeaza_distributia_bancnotelor(suma, bancnote = [10, 5, 2, 1]) -> pd.DataFrame:
    rezultat_dict = {

    }
    rest = suma
    for ban in bancnote:
        numar = rest // ban
        rezultat_dict[ban] = numar
        rest = rest % ban
    df = pd.DataFrame(rezultat_dict.values(), index=rezultat_dict.keys(), columns=["Bancnote"])
    return df
        
```

```python
df = calculeaza_distributia_bancnotelor(128)
df
```

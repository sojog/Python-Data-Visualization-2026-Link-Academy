```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
def calculeaza_distributia_bancnotelor(suma, bancnote = [10, 5, 2, 1]) -> pd.DataFrame:
    rezultat_dict = {

    }
    rest = suma
    for ban in bancnote:
        numar = rest // ban
        rezultat_dict[ban] = numar
        rest = rest % ban
    df = pd.DataFrame(rezultat_dict.values(), index=rezultat_dict.keys(), columns=["Bancnote"])
    return df
        
```

```python
df = calculeaza_distributia_bancnotelor(128)
df
```

```python
SHOULD_SAVE = True
```

```python
figura, axa = plt.subplots()

def update_function(frame_number):
    # print("Frame number:", frame_number)

    current_df = df.iloc[:frame_number+1]
    pie_chart = axa.pie(current_df.Bancnote)
    return pie_chart

ani = animation.FuncAnimation(fig=figura, frames=df.index.size,  func=update_function, interval=500)


if SHOULD_SAVE:
    ani.save("bancnote.gif", writer="ffmpeg")
    plt.show()
else:
    plt.close()
    from IPython.display import HTML
    HTML(ani.to_jshtml())


```

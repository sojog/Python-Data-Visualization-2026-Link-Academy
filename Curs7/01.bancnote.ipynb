```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
suma = 128
```

```python
bancnote = [10, 5, 2, 1]
bancnote
```

```python
rest = suma
```

```python
rezultat_dict = {

}

for ban in bancnote:
    numar = rest // ban
    print("numar:", numar)
    rezultat_dict[ban] = numar
    rest = rest % ban
```

```python
rezultat_dict
```

```python
df = pd.DataFrame(rezultat_dict.values(), index=rezultat_dict.keys(), columns=["Bancnote"])
df
```

```python
plt.pie(df.Bancnote, labels=df.index, autopct="%1.1f%%");
```

```python
plt.pie(rezultat_dict.values(), labels=rezultat_dict.keys(), autopct="%1.1f%%");
```

```python
plt.bar(df.index, df.Bancnote)
```

```python
plt.bar(list(df.index), list(df.Bancnote))
```

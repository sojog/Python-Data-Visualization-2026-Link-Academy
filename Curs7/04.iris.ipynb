# Scatter plot - inseamna relatia intre doua 2 variabile

```python
## 2D
plt.scatter(iris_df[sepal_length], iris_df[sepal_width])
```

```python
## 2D + Culoarea
plt.scatter(iris_df[sepal_length], iris_df[sepal_width], c=iris_df.species.map({"setosa":0, "versicolor":1, "virginica":2}));
```

```python
## 2D + Culoarea
sns.scatterplot(iris_df, x="sepal_length", y="sepal_width", hue="species");
```

```python
sns.scatterplot(iris_df, x="petal_length", y="petal_width", hue="species");
```

```python
sns.pairplot(iris_df, hue="species")
```

```python
numeric_columns = iris_df.keys()[:-1]
for feature in numeric_columns:
    sns.violinplot(iris_df, y=feature, x="species")
```

```python
numeric_columns = iris_df.keys()[:-1]
for feature in numeric_columns:
    sns.violinplot(iris_df, y=feature, x="species")
    plt.show()
```

```python
iris_df.corr(numeric_only=True) # Default Pearson
```

```python
iris_df.corr(method="spearman", numeric_only=True)
```

```python
correlation_matrix = iris_df.corr(method="pearson", numeric_only=True)

sns.heatmap(correlation_matrix, cmap="coolwarm")
```

```python
correlation_matrix = iris_df.corr(method="pearson", numeric_only=True)
sns.heatmap(correlation_matrix, cmap="coolwarm", vmin=-1, vmax=1)
```

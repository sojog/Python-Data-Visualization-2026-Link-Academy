## Formula Standard Deviation

### 1. Listare date

```python
data = [5, 8, 9, 12, 6, 10, 8, 6]
# data = [5, 5, 5, 5, 5, 5, 5, 5]
data
```

### 2. Media (average)

```python
avg = sum(data)  / len(data)
avg
```

```python
s_data = pd.Series(data)
```

```python
plt.scatter(s_data.index, s_data.values)
plt.axhline(avg)
```

### 3. Diferentele intre valori si media dar la patrat

```python
for i in data:
    print(f"{i} - {avg} = {i - avg}")
```

```python
col2 = [f"{i} - {avg} = {i - avg}" for i in data]
col3 = [f" {i - avg} ** 2 = {(i - avg) ** 2}" for i in data]
```

```python
pd.DataFrame({"data":data, "Patrat": col2, "Rezultat": col3})
```

### 4. Suma diferentelor la patrat

```python

desc = ""
suma = 0

for i in data:
    suma += (i - avg) ** 2
    desc += f"+ {(i-avg) ** 2}"

desc +=  f" = {suma}"
desc
```

### 5. Calcul varianță ( variance)

```python
variance = suma / len(data) # suma aici fiind suma patratelor
variance
```

```python
np.var(data)
```

# 6. Standard variance ( Deviația standard)

```python
np.sqrt(variance)
```

```python
np.std(data)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("World_Cup.csv")
#1
df.head(5)
```

```python
## 1. Creează un grafic bară care afișează numărul de trofee pentru fiecare echipă.

winners = df["Winner"].value_counts()
winners
```

```python
type(winners)
```

```python
plt.bar(winners.index, winners.values)
plt.xticks(rotation=45);
```

```python
plt.bar(winners.index, winners.values)
plt.xticks(rotation=45);
plt.axhline(winners.mean());
```
